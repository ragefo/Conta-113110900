```python
import pandas as pd
pd.__version__
```

```python
# Importar o arquivo gerado no Tesouro Gerencial
df_saldo = pd.read_excel('/content/Saldos Conta 113110900.xlsx')

# Renomear colunas
df_saldo.rename(columns={'Unnamed: 1':'UF Nome', 'Unnamed: 3':'UG Nome', 'Unnamed: 5':'Conta Contabil Nome', 'Mês Lançamento': 'Conta_Corrente', 'Unnamed: 7':'Entidade', 'UGE - UF':'UF'}, inplace=True)

# Excluir a linha 1
df_saldo = df_saldo.drop(0).reset_index(drop=True)

# Substituir os valores nulos por 0
df_saldo = df_saldo.infer_objects(copy=False)

# Substituir os espaços por _ do cabeçalho. Isso evita erros porque a linguagem pode apresentar erros ao lidar com espaços e acentos
df_saldo.columns = df_saldo.columns.str.replace(' ', '_')
df_saldo.columns = df_saldo.columns.str.replace('á', 'a')

# Definir a UG Executora como string
df_saldo['UG_Executora'] = df_saldo['UG_Executora'].astype(str).str.split('.').str[0]

# Criar a coluna Diferenca usando os índices das colunas para realizar a subtração. Isso evita que seja necessário usar o cabeçalho dos meses.
df_saldo['Diferenca'] = df_saldo.iloc[:, 8] - df_saldo.iloc[:, 9]

# Retorna apenas as linhas que apresentam Diferenca = 0 e o saldo no mês != 0
df_saldo = df_saldo[(df_saldo['Diferenca'] == 0) & (df_saldo.iloc[:, 8] != 0)].reset_index(drop=True)

# Mostra a planilha após as manipulações
df_saldo
```
