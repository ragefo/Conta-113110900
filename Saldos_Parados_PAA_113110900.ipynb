```python
# Importar o arquivo gerado no Pentaho SIGPAA
df_saldo_paa = pd.read_excel('/content/drive/MyDrive/Conab/Saldos 113110900/Saldos parados 113110900.xlsx')

# Renomear colunas
df_saldo_paa.rename(columns={'Mês Lançamento': 'Conta_Corrente', 'Unnamed: 4':'Entidade', 'UGE - UF':'UF'}, inplace=True)

# Excluir a linha 1
df_saldo_paa = df_saldo_paa.drop(0).reset_index(drop=True)

# Substituir os valores nulos por 0
df_saldo_paa = df_saldo_paa.infer_objects(copy=False)

# Substituir os espaços por _ do cabeçalho. Isso evita erros porque a linguagem pode apresentar erros ao lidar com espaços e acentos
df_saldo_paa.columns = df_saldo_paa.columns.str.replace(' ', '_')

# Definir a UG Executora como string
df_saldo_paa['UG_Executora'] = df_saldo_paa['UG_Executora'].astype(str).str.split('.').str[0]
df_saldo_paa['Conta_Contábil'] = df_saldo_paa['Conta_Contábil'].astype(str).str.split('.').str[0]

# Criar a coluna Diferenca usando os índices das colunas para realizar a subtração. Isso evita que seja necessário usar o cabeçalho dos meses.
df_saldo_paa['Diferenca'] = df_saldo_paa.iloc[:, 5] - df_saldo_paa.iloc[:, 6]

# Retorna apenas as linhas que apresentam Diferenca = 0 e o saldo no mês != 0
df_saldo_paa = df_saldo_paa[(df_saldo_paa['Diferenca'] == 0) & (df_saldo_paa.iloc[:, 5] != 0)].reset_index(drop=True)

# Mostra a planilha após as manipulações
df_saldo_paa
```
